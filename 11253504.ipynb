```python
import matplotlib.pyplot as mplt
import numpy as np
import pandas as pd
```

```python
data1 = pd.read_csv("red_wine.csv")
data1
```

```python
data1.plot(x = 'pH', y = 'density', xlabel = 'pH', ylabel = 'density', kind = 'scatter')
```

```python
data1.groupby('pH').count().plot()
mplt.savefig('plot1.png')
```

```python
data1.plot(x = 'alcohol', y = 'quality', xlabel = 'alcohol', ylabel = 'quality', kind = 'scatter')
data1.groupby('alcohol').count().plot()
mplt.savefig('plot2.png')
```

```python
data1.plot(x = 'citric acid', y = 'residual sugar', xlabel = 'citric acid', ylabel = 'residual sugar', kind = 'scatter')
data1.groupby('citric acid').count().plot()
mplt.savefig('plot3.png')
```

```python
data1.plot(x = 'fixed acidity', y = 'chlorides', xlabel = 'fixed acidity', ylabel = 'chlorides', kind = 'scatter')
data1.groupby('fixed acidity').count().plot()
mplt.savefig('plot4.png')
```
